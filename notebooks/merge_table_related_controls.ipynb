```python
import pandas as pd
```

## EXP 5 LAB

```python
exp5lab = pd.read_csv('../data/2023_FISH_VNIIPRH - EXP_5_main_lab.csv')
```

```python
exp5lab
```

```python
exp5lab.info()
```

```python
exp5lab.columns
```

```python
exp5lab.FEMALE = exp5lab.FEMALE.astype(str)
exp5lab.MALE = exp5lab.MALE.astype(str)
exp5lab['FAMILY'] = exp5lab.FEMALE + 'x' + exp5lab.MALE

exp5lab.CUP_NUMBER = exp5lab.CUP_NUMBER.astype(str)
```

```python
exp5lab_short = exp5lab.pivot(index=['FAMILY', 'MUT', 'CUP_NUMBER'], columns='TEMP', values=['FERT', 'HATCHED_ALIVE'])
exp5lab_short = exp5lab_short.reset_index()

exp5lab_short.columns = ['FAMILY', 'MUT', 'CUP_NUMBER', 'Aca', 'Aco', 'Hca', 'Hco']
exp5lab_short['TEMP'] = 38
exp5lab_short['TIME'] = 40
exp5lab_short = exp5lab_short[['FAMILY', 'CUP_NUMBER', 'MUT', 'TEMP', 'TIME', 'Aca', 'Hca', 'Aco', 'Hco']]

exp5lab_short['RelativeDecrease'] = (exp5lab_short.Hca / exp5lab_short.Aca) / (exp5lab_short.Hco / exp5lab_short.Aco)
```

```python
exp5lab_short
```

```python
exp5lab_short.to_csv('../data/2023_derived_tables/exp5lab.csv')
```

## EXP 5 FACTORY

```python
exp5fac = pd.read_csv('../data/FISH_TABLES/clear_EXP_5_main_factory.csv')
```

```python
exp5fac
```

```python
exp5fac.info()
```

```python
exp5fac.FEMALE = exp5fac.FEMALE.astype(str)
exp5fac.MALE = exp5fac.MALE.astype(str)
exp5fac['FAMILY'] = exp5fac.FEMALE + 'x' + exp5fac.MALE
```

```python
exp5fac_short = exp5fac.pivot(index=['FAMILY', 'MUT'], columns='TEMP', values=['FERT.1', 'HATCHED', 'SWIM'])
exp5fac_short = exp5fac_short.reset_index()

exp5fac_short.columns = ['FAMILY', 'MUT', 'Aca', 'Aco', 'Hca', 'Hco', 'Sca', 'Sco']
exp5fac_short['TEMP'] = 38
exp5fac_short['TIME'] = 40
exp5fac_short = exp5fac_short[['FAMILY', 'MUT', 'TEMP', 'TIME', 'Aca', 'Hca', 'Sca', 'Aco', 'Hco', 'Sco']]

exp5fac_short['RelativeDecreaseHatched'] = (exp5fac_short.Hca / exp5fac_short.Aca) / (exp5fac_short.Hco / exp5fac_short.Aco)
exp5fac_short['RelativeDecreaseSwim'] = (exp5fac_short.Sca / exp5fac_short.Aca) / (exp5fac_short.Sco / exp5fac_short.Aco)
```

```python
exp5fac_short
```

```python
exp5fac_short.to_csv('../data/2023_derived_tables/exp5factory.csv')
```
