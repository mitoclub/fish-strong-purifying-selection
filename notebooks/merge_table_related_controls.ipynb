## EXP 5 LAB

```python
sheet_id = '1aqHnDaiRXntyrbj6qaqzR-4XnLbvtKU84FB_JLu5x5A'
sheet_name = 'EXP_5_main_lab'
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}'
exp5lab = pd.read_csv(url)
```

```python
exp5lab
```

```python
exp5lab.info()
```

```python
exp5lab.columns
```

```python
exp5lab.FEMALE = exp5lab.FEMALE.astype(str)
exp5lab.MALE = exp5lab.MALE.astype(str)
exp5lab['FAMILY'] = exp5lab.FEMALE + 'x' + exp5lab.MALE

exp5lab.CUP_NUMBER = exp5lab.CUP_NUMBER.astype(str)
```

```python
exp5lab_short = exp5lab.pivot(index=['FAMILY', 'MUT', 'CUP_NUMBER'], columns='TEMP', values=['FERT', 'HATCHED_ALIVE', 'SWIM'])
exp5lab_short = exp5lab_short.reset_index()

exp5lab_short.columns = ['FAMILY', 'MUT', 'CUP_NUMBER', 'Aca', 'Aco', 'Hca', 'Hco', 'Sca', 'Sco']
exp5lab_short['TEMP'] = 38
exp5lab_short['TIME'] = 40
exp5lab_short = exp5lab_short[['FAMILY', 'CUP_NUMBER', 'MUT', 'TEMP', 'TIME', 'Aca', 'Hca', 'Aco', 'Hco', 'Sca', 'Sco']]
```

```python
exp5lab_short_sum = exp5lab_short.groupby(by=['FAMILY', 'MUT', 'TEMP', 'TIME']).sum()
exp5lab_short_sum = exp5lab_short_sum.reset_index()
```

```python
exp5lab_short['RelativeDecreaseHatched'] = (exp5lab_short.Hca / exp5lab_short.Aca) / (exp5lab_short.Hco / exp5lab_short.Aco)
exp5lab_short['RelativeDecreaseSwim'] = (exp5lab_short.Sca / exp5lab_short.Aca) / (exp5lab_short.Sco / exp5lab_short.Aco)

exp5lab_short_sum['RelativeDecreaseHatched'] = (exp5lab_short_sum.Hca / exp5lab_short_sum.Aca) / (exp5lab_short_sum.Hco / exp5lab_short_sum.Aco)
exp5lab_short_sum['RelativeDecreaseSwim'] = (exp5lab_short_sum.Sca / exp5lab_short_sum.Aca) / (exp5lab_short_sum.Sco / exp5lab_short_sum.Aco)
```

```python
exp5lab_short
```

```python
exp5lab_short_sum
```

```python
exp5lab_short.to_csv('../data/2023_derived_tables/exp5lab.csv')

exp5lab_short_sum.to_csv('../data/2023_derived_tables/exp5lab_sum.csv')
```
