```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
sheet_id = '1aqHnDaiRXntyrbj6qaqzR-4XnLbvtKU84FB_JLu5x5A'
sheet_name = 'EXP_8_second_main_factory'
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}'
exp8fac = pd.read_csv(url)
```

```python
exp8fac.TEMP = exp8fac.TEMP.astype(str)

exp8fac.FERT_PERC = exp8fac.FERT_PERC.str.replace(',', '.')
exp8fac.FERT_PERC = pd.to_numeric(exp8fac.FERT_PERC)

exp8fac.SWIM_PERC_FERT = exp8fac.SWIM_PERC_FERT.str.replace(',', '.')
exp8fac.SWIM_PERC_FERT = pd.to_numeric(exp8fac.SWIM_PERC_FERT)

exp8fac.SWIM_PERC_FERT_NORM = exp8fac.SWIM_PERC_FERT_NORM.str.replace(',', '.')
exp8fac.SWIM_PERC_FERT_NORM = pd.to_numeric(exp8fac.SWIM_PERC_FERT_NORM)
```

```python
exp8fac = exp8fac.iloc[9:13,].reset_index(drop=True)
```

```python
exp8fac
```

```python
code = []
mut = []
temp = []
time = []
fert = []
swim = []

for row in exp8fac.index:
    total_count = int(exp8fac.loc[row, 'TOTAL'])
    fert_count = int(exp8fac.loc[row, 'FERT.1'])
    swim_count = int(exp8fac.loc[row, 'SWIM'])
    
    code += [exp8fac.loc[row, 'CODE']] * total_count
    mut += [exp8fac.loc[row, 'MUT']] * total_count
    temp += [exp8fac.loc[row, 'TEMP']] * total_count
    time += [exp8fac.loc[row, 'TIME']] * total_count
    
    fert += [1] * fert_count
    fert += [0] * (total_count - fert_count)
    swim += [1] * swim_count
    swim += [0] * (total_count - swim_count)
```

```python
ind_data = pd.DataFrame({'CODE': code, 'MUT': mut, 'TEMP': temp, 'TIME': time, 'FERT': fert, 'SWIM': swim})
ind_data.head()
```

```python
# НЕ ЗАПУСКАТЬ!!!
bs_cc = st.bootstrap((ind_data[ind_data['CODE'] == 'CC']['SWIM'], ), np.count_nonzero, n_resamples=1000)
bs_c38 = st.bootstrap((ind_data[ind_data['CODE'] == 'C38']['SWIM'], ), np.count_nonzero, n_resamples=1000)
bs_mc = st.bootstrap((ind_data[ind_data['CODE'] == 'MC']['SWIM'], ), np.count_nonzero, n_resamples=1000)
bs_m38 = st.bootstrap((ind_data[ind_data['CODE'] == 'M38']['SWIM'], ), np.count_nonzero, n_resamples=1000)
```

```python
# доверительный интервал для доли
exp8fac['CI_upper'] = (exp8fac.SWIM_PERC_FERT/100) +      \
                        (np.sqrt((exp8fac.SWIM_PERC_FERT/100) * (1 - (exp8fac.SWIM_PERC_FERT/100)) / \
                        exp8fac['FERT.1']) * 1.96)
exp8fac['CI_lower'] = (exp8fac.SWIM_PERC_FERT/100) -      \
                        (np.sqrt((exp8fac.SWIM_PERC_FERT/100) * (1 - (exp8fac.SWIM_PERC_FERT/100)) / \
                        exp8fac['FERT.1']) * 1.96)
```

```python
# доверительный интервал для доли нормированной
exp8fac['CI_upper_norm'] = (exp8fac.SWIM_PERC_FERT_NORM/100) +      \
                        (np.sqrt((exp8fac.SWIM_PERC_FERT_NORM/100) * (1 - (exp8fac.SWIM_PERC_FERT_NORM/100)) / \
                        exp8fac['FERT.1']) * 1.96)
exp8fac['CI_lower_norm'] = (exp8fac.SWIM_PERC_FERT_NORM/100) -      \
                        (np.sqrt((exp8fac.SWIM_PERC_FERT_NORM/100) * (1 - (exp8fac.SWIM_PERC_FERT_NORM/100)) / \
                        exp8fac['FERT.1']) * 1.96)
```

```python
exp8fac['CI'] = np.sqrt((exp8fac.SWIM_PERC_FERT/100) * (1 - (exp8fac.SWIM_PERC_FERT/100)) / exp8fac['FERT.1']) * 1.96
exp8fac['CI_norm'] = np.sqrt((exp8fac.SWIM_PERC_FERT_NORM/100) * (1 - (exp8fac.SWIM_PERC_FERT_NORM/100)) / exp8fac['FERT.1']) * 1.96
```

```python
exp8fac
```

```python
exp8fac['CODE_COND'] = exp8fac.CODE.str.replace('1', '').str.replace('2', '')
```

```python
# expected line logic: (shock / control) * (mutagen / control) * control
expected_8 = (exp8fac[exp8fac['CODE_COND'] == 'C38'].SWIM_PERC_FERT_NORM.mean() / exp8fac[exp8fac['CODE_COND'] == 'CC'].SWIM_PERC_FERT_NORM.mean()) * \
         (exp8fac[exp8fac['CODE_COND'] == 'MC'].SWIM_PERC_FERT_NORM.mean() / exp8fac[exp8fac['CODE_COND'] == 'CC'].SWIM_PERC_FERT_NORM.mean()) * \
          exp8fac[exp8fac['CODE_COND'] == 'CC'].SWIM_PERC_FERT_NORM.mean()
```

```python
sns.barplot(exp8fac, x='CODE_COND', y='SWIM_PERC_FERT_NORM', yerr=exp8fac['CI_norm'])
plt.axhline(y=expected_8, xmin=0.75, color='orange', ls='--');
```
